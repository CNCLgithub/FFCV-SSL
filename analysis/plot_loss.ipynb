```python
import json
import pandas as pd
import seaborn as sns
```

```python
data = []
with open('../logs-monkey/log', 'r') as file:
    for line in file:
        data.append(json.loads(line))
```

```python
data = pd.DataFrame.from_records(data) 
```

```python
data.describe()
print(data.columns)
data[["train_loss", "loss_classif_val_layer0"]]
```

```python
sns.lineplot(data=data, x="epoch", y="train_loss")
```
